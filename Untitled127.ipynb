```python
import pandas as pd
import torch
from diffusers import DiffusionPipeline
from torch.utils.data import Dataset, DataLoader
from transformers import CLIPTokenizer
from transformers import AutoTokenizer, CLIPTextModel
from torchvision import transforms
from PIL import Image
import os

# Load the dataset
file_path = "/content/verse_dataset (4).csv"
dataset = pd.read_csv(file_path)


class VerseImageDataset(Dataset):
    def __init__(self, dataframe, tokenizer, image_folder="images"): # Changed _init_ to __init__
        self.data = dataframe
        self.tokenizer = tokenizer
        self.image_folder = image_folder
        self.transform = transforms.Compose([
            transforms.Resize((512, 512)),
            transforms.ToTensor(),
            transforms.Normalize([0.5], [0.5]),  # Normalize to [-1, 1] range
        ])

    def __len__(self): # Changed _len_ to __len__
        return len(self.data)

    def __getitem__(self, idx): # Changed _getitem_ to __getitem__
        row = self.data.iloc[idx]
        prompt = row["Verse Text"]
        # Padding and truncation should happen during batch creation in the collate_fn
        tokens = self.tokenizer(prompt, return_tensors="pt", padding=False, truncation=False)
        image_path = os.path.join(self.image_folder, f"{row['Verse Num']}.jpg")
        if os.path.exists(image_path):
            image = Image.open(image_path).convert("RGB")
            image = self.transform(image)
        else:
            image = torch.zeros((3, 512, 512))  # Placeholder if no image is available
        return tokens.input_ids, image



 #Initialize tokenizer
tokenizer = CLIPTokenizer.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", subfolder="tokenizer")


# Prepare dataset and dataloader
# Define a custom collate function to handle variable-length input_ids
def collate_fn(batch):
    input_ids_list, images_list = zip(*batch)

    # Pad input_ids to the maximum length in the batch
    max_len = max(input_ids.shape[1] for input_ids in input_ids_list)
    padded_input_ids = [torch.nn.functional.pad(input_ids, (0, max_len - input_ids.shape[1]), value=tokenizer.pad_token_id) for input_ids in input_ids_list]

    # Stack padded input_ids and images
    input_ids = torch.stack(padded_input_ids)
    images = torch.stack(images_list)

    return input_ids, images

image_folder = "/content/image_folder"  # Replace with the path to your images folder
dataset = VerseImageDataset(dataset, tokenizer, image_folder)
# Use the custom collate function in the DataLoader
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)

# Load the Stable Diffusion XL pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
).to("cuda")





# Function to generate images
def generate_image(prompt):
    images = pipe(prompt=prompt).images[0]
    images.save(f"{prompt[:50]}.png")
    return images

# Iterate through the underlying DataFrame within the dataset object
for idx in range(len(dataset)):
    row = dataset.data.iloc[idx]  # Access the DataFrame row using self.data
    prompt = row["Verse Text"]
    image = generate_image(prompt)
    image.show()
```
